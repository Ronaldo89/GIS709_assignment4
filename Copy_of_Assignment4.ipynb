```python
##Import libraries necessary for program
import requests
import pandas as pd
import json
from datetime import datetime,timedelta
```

```python
##Link to earthquake data and parameters
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2018-01-01&endtime=2018-02-02&minmagnitude=5'
r = requests.get(url) 

```

```python
##Extract Data from the Request using the Json() Method
data_request = json.loads(r.text)
print(data_request)
data_request.keys() 
```

```python
##Normalize to dataframe 
USGS_eq = pd.json_normalize(data_request, 'features')
USGS_eq.head(1)
```

```python
##clean the data 
USGS_eq.rename(columns = {'properties.mag':'Magnitude'},inplace = True)
USGS_eq.rename(columns = {'properties.place':'Place'},inplace = True)
USGS_eq.rename(columns = {'properties.time':'Time'},inplace = True)
USGS_eq.rename(columns = {'properties.updated':'Updated Time'},inplace = True)
USGS_eq.rename(columns = {'properties.url':'Url'},inplace = True)
USGS_eq.rename(columns = {'properties.detail':'Details'},inplace = True)
USGS_eq.rename(columns = {'properties.felt':'Reported'},inplace = True)
USGS_eq.rename(columns = {'properties.types':'Property Type'},inplace = True)
USGS_eq.rename(columns = {'properties.nst':'Nst'},inplace = True)
USGS_eq.rename(columns = {'properties.dmin':'Admin'},inplace = True)
USGS_eq.rename(columns = {'properties.rms':'Rms'},inplace = True)
USGS_eq.rename(columns = {'properties.gap':'Gap'},inplace = True)
USGS_eq.rename(columns = {'properties.magType':'Mag Type'},inplace = True)
USGS_eq.rename(columns = {'properties.types':'Type'},inplace = True)
USGS_eq.rename(columns = {'properties.title':'Properties Title'},inplace = True)
USGS_eq.rename(columns = {'geometry.type':'Geometry Type'},inplace = True)
USGS_eq.rename(columns = {'geomerty.coordinates':'Coordinates'},inplace = True)


USGS_eq.head(2)
```

```python
##Split coordinates into lat, long and depth
USGS_eq[['Latitude','Longitude','Depth']]= USGS_eq['geometry.coordinates'].apply(pd.Series) 
USGS_eq.head(2)
```

```python
# converting the integers to datetime format
USGS_eq['Time'] = pd.to_datetime(USGS_eq['Time'], utc = True, unit = 'ms')
USGS_eq['Updated'] = pd.to_datetime(USGS_eq['Updated Time'], utc= True, unit = 'ms')
USGS_eq.head(1)
```

```python
#Extract cleaned column
USGC_catalogue = USGS_eq[['id', 'Magnitude', 'Place', 'Time', 'Updated Time', 'Reported', 'Latitude', 'Longitude', 'Depth']].copy()
USGC_catalogue.head(6)
```

```python
##Create Map of the data
import folium
map = folium.Map(location=[USGC_catalogue.Latitude.mean(), USGC_catalogue.Longitude.mean()] ,max_bounds= True,zoom_start=5, control_scale=True,  tiles="Stamen Terrain")
for index, location_info in USGC_catalogue.iterrows():
    folium.Marker([location_info['Longitude'], location_info["Latitude"]],tooltip = 'Check earthquake parameters',popup=location_info["Place"]).add_to(map)

##Visualizing historical seismic activity
def seismic(self):
    seismic_data = pd.readcsv('https://www.kaggle.com/datasets/usgs/earthquake-database?resource=download&select=database.csv')
    #extracting coordinates from data set
    lon = []
    for i in seismic_data['Longitude']:
        lon.append(i)
    lat = []
    for i in seismic_data['Latitude']:
        lat.append(i)
       
    ##Plotting historical data
    for x in range(len(lon)):
        folium.Choropleth(location=[lat[x],lon[x]],
                          geo_data=seismic_data,
                          name="choropleth",
                          data=seismic_data,
                          columns=["Maginitude"],
                          fill_color="YlGn",
                          legend_name="Seismic Zones",
                          ).add_to(map)
map
```
